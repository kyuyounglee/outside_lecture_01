```python
import pandas as pd

df = pd.read_csv("nasdaq_screener.csv", na_filter=False)
df.head()
```

```python
df[df['Symbol'].str.contains(r'[/^]', '_', regex=True)]
```

```python
df['id'] = df['Symbol'].str.strip().replace(r'[/^]', '_', regex=True)
df.head()
```

```python
result_d = df.to_dict(orient="records")
result_d
```

```python
from dotenv import load_dotenv
import os
import meilisearch

load_dotenv()

client  = meilisearch.Client(url=os.getenv("MEILISEARCH_URL"), api_key=os.getenv("MEILISEARCH_API_KEY"))
```

```python
client.index("nsdaq").add_documents(result_d, primary_key='id')
```

```python
client.index('nsdaq').search('Apple inc')
```
